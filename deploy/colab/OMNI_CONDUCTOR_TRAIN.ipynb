```python
from huggingface_hub import login
from google.colab import userdata
import os

token = userdata.get('HF_TOKEN')
login(token=token)
os.environ['HF_TOKEN'] = token
print('HF login OK')
```

```python
import urllib.request
from pathlib import Path

DATA_URL = (
    'https://raw.githubusercontent.com/ujjwal-coder211/Saas/main/'
    'omni_training/data/export/conductor_v1_train.jsonl'
)
DATA_PATH = Path('/content/conductor_v1_train.jsonl')
urllib.request.urlretrieve(DATA_URL, DATA_PATH)
print('Downloaded', DATA_PATH, 'bytes', DATA_PATH.stat().st_size)
```

```python
from unsloth import FastLanguageModel

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name='nvidia/Nemotron-3-Nano-30B-A3B',
    max_seq_length=4096,
    load_in_4bit=True,
    dtype=None,
)

model = FastLanguageModel.get_peft_model(
    model,
    r=32,
    lora_alpha=64,
    target_modules=['q_proj','k_proj','v_proj','o_proj','gate_proj','up_proj','down_proj'],
    lora_dropout=0.05,
    bias='none',
    use_gradient_checkpointing=True,
)
print('Model + LoRA ready')
```

```python
import json
from datasets import Dataset

rows = []
with open('/content/conductor_v1_train.jsonl', encoding='utf-8') as f:
    for line in f:
        if line.strip():
            rows.append(json.loads(line))

def format_example(ex):
    return tokenizer.apply_chat_template(ex['messages'], tokenize=False, add_generation_prompt=False)

dataset = Dataset.from_list([{'text': format_example(r)} for r in rows])
print('Training rows:', len(dataset))
```

```python
from trl import SFTTrainer, SFTConfig

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset,
    args=SFTConfig(
        output_dir='/content/omni_conductor_lora',
        per_device_train_batch_size=2,
        gradient_accumulation_steps=8,
        num_train_epochs=3,
        learning_rate=2e-4,
        lr_scheduler_type='cosine',
        warmup_ratio=0.1,
        logging_steps=10,
        save_steps=200,
        fp16=True,
        dataset_text_field='text',
    ),
)
trainer.train()
print('Training complete')
```
